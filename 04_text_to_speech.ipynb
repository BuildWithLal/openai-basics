```python
import os
import IPython
from openai import OpenAI
```

```python
client = OpenAI()
```

```python
response = client.audio.speech.create(
  model="tts-1",
  voice="alloy",
  input="OpenAI is great for turning text into speech."
)
```

```python
speech_file_path = os.path.abspath('speech.mp3')
```

```python
response.stream_to_file(speech_file_path)
```

```python
IPython.display.Audio("speech.mp3")
```

### Supported output formats

```python
response = client.audio.speech.create(
  model="tts-1",
  voice="nova",
  input="OpenAI is great for turning text into speech.",
  response_format="aac"  
)
```

```python
speech_file_path = os.path.abspath('output.aac')
```

```python
response.stream_to_file(speech_file_path)
```

```python
IPython.display.Audio("output.aac")
```
