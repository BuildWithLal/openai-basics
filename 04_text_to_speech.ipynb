```python
import os
import IPython
from openai import OpenAI
```

```python
client = OpenAI()
```

```python
speech_file_path = os.path.abspath('speech.mp3')

response = client.audio.speech.create(
  model="tts-1",
  voice="alloy",
  input="OpenAI is great for turing text into speech."
)

response.stream_to_file(speech_file_path)
```

```python
IPython.display.Audio("speech.mp3")
```

# 

### Audio Speed

```python
speech_file_path = os.path.abspath('speech-with-speed.mp3')

response = client.audio.speech.create(
  model="tts-1",
  voice="alloy",
  input="OpenAI is great for turing text into speech. it has different voice options and can increase or decrease audio speed as well",
  speed="2" # 0.25 to 4.0. default is 1.0
)

response.stream_to_file(speech_file_path)
```

```python
IPython.display.Audio("speech-with-speed.mp3")
```

#

### Streaming real time audio

```python
response = client.audio.speech.create(
    model="tts-1",
    voice="alloy",
    input="OpenAI is great for turing text into speech and playing in real time.",
)

response.stream_to_file("real-time.mp3")
```

```python
IPython.display.Audio("real-time.mp3")
```

#

### Supported output formats

```python
response = client.audio.speech.create(
    model="tts-1",
    voice="alloy",
    response_format="aac",
    input="OpenAI TTS supports couple of formats for audio output.",
)

response.stream_to_file("different-format.aac")
```

```python
IPython.display.Audio("real-time.mp3")
```
